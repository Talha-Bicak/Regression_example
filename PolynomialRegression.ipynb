```python
#Kütüphanelerin import edilmesi
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Veri setimizi oluşturmak için kullanılacak method
def pol(x):
    return x*np.cos(x)
```

```python
x = np.linspace(0,10,80)
np.random.seed(1234)
```

```python
y = pol(x) + np.random.randn(80)*2
```

```python
#DataFrame oluşturma
df = pd.DataFrame({'x':x,'y':y})
```

```python
#DataFrame'i kullanarak bir csv dosyası oluşturma
df.to_csv('data_set.csv', index=False)
```

```python
#Oluşturulan csv dosyasını okuma işlemi
df = pd.read_csv('data_set.csv')
df.head()
```

```python
#Saçılım grafiği
plt.scatter(df['x'],df['y'],color = 'green',s=100,marker='.')
```

```python
X= df.iloc[:,0:1].values
Y= df.iloc[:,1:2].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Linear Regression Modeli 
lin_reg=LinearRegression()
lin_reg.fit(X, Y)
plt.scatter(X, Y, color='red')
plt.title("Linear Regression")
plt.plot(x, lin_reg.predict(X),color='blue')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
#Polinomal Regresyon Modeli 
for degree in [2,4,6]:
    poly_reg = PolynomialFeatures(degree) #for döngüsü kullanarak 2-4-6 derecelerde modelin başarısı ölçülmüştür.
    x_poly = poly_reg.fit_transform(X)
    lin_reg2= LinearRegression()
    lin_reg2.fit(x_poly,y)
    plt.scatter(X,Y,color='red')
    plt.title("Polynomial Regresion,Degree: {}".format(degree))
    plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)),color="blue")
    plt.show()
```
